### Convert datatypes

```python
# Ensure 'discounted_price' and 'actual_price' are strings before applying replacements
data['discounted_price'] = data['discounted_price'].astype(str).str.replace("₹", '').str.replace(",", '')
data['actual_price'] = data['actual_price'].astype(str).str.replace("₹", '').str.replace(",", '')

# Convert 'discounted_price' and 'actual_price' to float after cleaning the strings
data['discounted_price'] = data['discounted_price'].astype(float)
data['actual_price'] = data['actual_price'].astype(float)

# Ensure 'discount_percentage' is a string before removing '%'
data['discount_percentage'] = data['discount_percentage'].astype(str).str.replace("%", '')
data['discount_percentage'] = data['discount_percentage'].astype(float)

# Convert 'rating_count' to integer, handle any NaNs or non-numeric values if necessary
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce').fillna(0).astype(float)
```

```python
data['rating'].unique()
```

### Impute null values with Median

```python
data['rating'] = pd.to_numeric(data['rating'].replace('|','4.1'))
data['rating'].fillna(data['rating'].median(), inplace=True)
```

### Initial Data Exploration/Visualization

```python
numeric_columns = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
```

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numeric_columns):
    plt.subplot(3, 2, i + 1)  # 3 rows, 2 columns
    sns.histplot(data[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data[numeric_columns])
plt.show()
```

```python
# Box plot for each numeric column
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numeric_columns):
    plt.subplot(3, 2, i + 1)  # Adjust number of rows and columns if necessary
    sns.boxplot(y=data[feature])
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

```python
data[numeric_columns].skew()
```

```python
# Detect outliers using IQR
X = data[numeric_columns]

# Calculate Q1 and Q3
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((X < lower_bound) | (X > upper_bound)).any(axis=1)
outliers_df = data[outliers]

print(f'Number of outliers detected: {outliers_df.shape[0]}')
```

```python
data_new = data.copy()
# Winsorize outliers
for col in numeric_columns:
    data_new.loc[data_new[col] < lower_bound[col], col] = lower_bound[col]
    data_new.loc[data_new[col] > upper_bound[col], col] = upper_bound[col]
```

```python
data_new[numeric_columns].skew()
```

```python
# Drop rating_count column
data_new.drop('rating_count', axis=1, inplace=True)
```
